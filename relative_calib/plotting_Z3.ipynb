```python
'''

    CDE Side - 1

'''
filename_1 = 'reproc_v2.1.0_Z3_11ser_check_norm.root'
filename_2 = 'reproc_v2.1.1_Z3_11ser_check_norm.root'
ltext = ['reproc_v2.1.0','reproc_v2.1.1']

file1 = rt.TFile.Open(filename_1)
file2 = rt.TFile.Open(filename_2)
gr1 = 'CDE_S1_1kev_corr'
gr2 = 'CDE_S1_1kev_corr'

graphs = []
graphs.append(file1.Get(gr1))
graphs.append(file2.Get(gr2))

legend = rt.TLegend(0.7,0.9,0.9,0.8)
legend.SetTextSize(0.035)
lentry = []
#ltext = ['Prompt','Reporcessed']
#ltext = ['Reporcessed','v1']


can = rt.TCanvas()
color = [rt.kBlack, rt.kRed]

title = 'Around 1.3 keV line (S1) C-D-E'
for i,g in enumerate(graphs):
    g.SetMarkerStyle(rt.kFullDotMedium)
    g.SetMarkerColor(color[i])
    #gr.SetMarkerStyle(rt.kCircle)
    g.GetXaxis().SetTitle('X')
    g.GetYaxis().SetTitle('Y')
    g.SetTitle(title)
    if(i == 0):
        g.Draw('AP*')
    else:
        g.Draw('P* same')
    lentry.append(legend.AddEntry(g,ltext[i],'p'))
    lentry[i].SetTextColor(g.GetMarkerColor())

linecolor = rt.kBlue
lE1 = rt.TLine(0., 0., 1.0, 0.)
lE1.SetLineColor(linecolor)
lE1.SetLineStyle(2)
lE1.SetLineWidth(2)
lE1.Draw()

lC1 = rt.TLine(0., 0., -0.5, 1.0)
lC1.SetLineColor(linecolor)
lC1.SetLineStyle(2)
lC1.SetLineWidth(2)
lC1.Draw()

lD1 = rt.TLine(0., 0., -0.5, -1.0)
lD1.SetLineColor(linecolor)
lD1.SetLineStyle(2)
lD1.SetLineWidth(2)
lD1.Draw()

latex = rt.TLatex()
latex.SetTextColor(linecolor)
latex.DrawLatexNDC(0.15,0.85,'#bf{C1}')
latex.DrawLatexNDC(0.15,0.15,'#bf{D1}')
latex.DrawLatexNDC(0.85,0.55,'#bf{E1}')

can.SetGridx()
can.SetGridy()
legend.Draw()

graphs[0].GetXaxis().SetLimits(-0.4,0.6)
graphs[0].GetYaxis().SetRangeUser(-0.5,0.55)

can.Draw()
if(save):
    savename='CDE_S1_1kev_Z3.png'
    can.SaveAs('plots/%s'%(savename))
```
